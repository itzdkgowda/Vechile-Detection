```python
import cv2
from ultralytics import YOLO

# Load YOLOv8 model
model = YOLO("yolov8n.pt")

# Open camera (or use a video)
cap = cv2.VideoCapture(0)  # Change to 'video.mp4' for a recorded video

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO detection
    results = model(frame)
    
    vehicle_detected = False
    
    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            label = result.names[int(box.cls[0])]

            # Detect only vehicles (cars, trucks, buses, motorcycles)
            if label in ["car", "truck", "bus", "motorcycle"]:
                vehicle_detected = True
                break
    
    # Create an overlay
    overlay = frame.copy()
    color = (0, 255, 0)  # Default to green (safe)
    text = "SAFE"

    if vehicle_detected:
        color = (0, 0, 255)  # Red overlay (stop)
        text = "STOP!"
    
    cv2.rectangle(overlay, (0, 0), (frame.shape[1], frame.shape[0]), color, -1)
    alpha = 0.4  # Transparency level
    frame = cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0)
    
    # Display text
    cv2.putText(frame, text, (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 5)
    
    # Show the frame
    cv2.imshow("Vehicle Detection", frame)
    
    # Exit on 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```
